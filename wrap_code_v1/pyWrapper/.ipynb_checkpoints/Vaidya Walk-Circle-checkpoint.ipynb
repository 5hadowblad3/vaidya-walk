```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from basics import *
import main_run
```

```python
def line(k, N):
    y0, x0 = np.sin(2 * np.pi * (k)/N), np.cos(2 * np.pi * (k)/N)
    y1, x1 = np.sin(2 * np.pi * (k+1)/N), np.cos(2 * np.pi * (k+1)/N)
    m = (y1-y0)/(x1-x0)
    c = y0 - m * x0
    return m, c

def a_and_b(k, N):
    m, c = line(k, N)
    a = np.array([1., -m])
    if c > 0:
        b = c
    else:
        a *= -1
        b = -c
    return a, b

def generate_A_and_b(N, d=2):
    A = np.zeros((N, d))
    b = np.zeros(N)
    for k in range(N):
        A[k, :], b[k] = a_and_b(k, N)
    return A, b

def plot_one_line(a, b, xx, xlims, ylims):
    plt.plot(xx, (b-a[0]*xx)/a[1], linewidth=1.)
    axes = plt.gca()
    axes.set_xlim(xlims)
    axes.set_ylim(ylims)
    axes.grid('on')

def plot_A_b(A, b, xx, xlims, ylims):
    n, d = A.shape
    for i in range(n):
        plt.plot(xx, (b[i]-A[i, 0]*xx)/A[i, 1], linewidth=1.)
    axes = plt.gca()
    axes.set_xlim(xlims)
    axes.set_ylim(ylims)
    axes.grid('on')
    plt.axhline(0, color='black', linewidth=.5)
    plt.axvline(0, color='black', linewidth=.5)
    circle1 = plt.Circle((0, 0), 1, color='black', fill=False)
    axes.add_artist(circle1)
#     plt.show()
```

```python
np.random.seed(695829)
experiments = 100
iterations = 80

# Defining the Polytope
d = 2
r = 0.5
As = []
bs = []

Ns = [4, 16, 260]
Nvals = len(Ns)
rvs_rep = np.zeros((Nvals, d, experiments, iterations))
rvs2_rep = np.zeros((Nvals, d, experiments, iterations))
rvs3_rep = np.zeros((Nvals, d, experiments, iterations))
for i, N in enumerate(Ns):
    print 'N = %d ' % N
    A, b = generate_A_and_b(N)
    As.append(A)
    bs.append(b)
    rvs_rep[i,:, :, :] = main_run.run_dikin_walk(experiments, iterations, A, b, r)
    rvs2_rep[i,:, :, :] = main_run.run_vaidya_walk(experiments, iterations, A, b, r)
    rvs3_rep[i,:, :, :] = main_run.run_john_walk(experiments, iterations, A, b, r)
    
```

```python
fig = plt.figure(figsize = [5, 5])

N = 5
xlims = np.array([-2., 2])
ylims = np.array([-2, 2])
xx = np.linspace(xlims[0], xlims[1], 21)

A, b = generate_A_and_b(N)
plot_A_b(A, b, xx, xlims, ylims)
# plt.show()
```

```python
I = 4
fig = plt.figure(figsize = [I*3, 3*Nvals*3.5])

for k in range(Nvals):
    for j, i in enumerate(range(0, iterations, iterations/(I-1))):
        A = As[k]
        b = bs[k]
        n = len(b)
        
        x = rvs_rep[k, 0, :, i]
        y = rvs_rep[k, 1, :, i]
        ax = fig.add_subplot(3*Nvals,I,(3*k)*I+j+1)
        ax.scatter(x, y, color='b')
        plot_A_b(A, b, xx, xlims, ylims)
        axes = plt.gca()
        axes.set_xlim([-1.,1.])
        axes.set_ylim([-1.,1.])
        axes.set_xticks([])
        axes.set_yticks([])
        plt.title('D, iter = %d, m=%d' % (i,n))

        x = rvs2_rep[k, 0, :, i]
        y = rvs2_rep[k, 1, :, i]
        ax = fig.add_subplot(3*Nvals,I,(3*k+1)*I+j+1)
        ax.scatter(x, y,color='r')
        plot_A_b(A, b, xx, xlims, ylims)
        axes = plt.gca()
        axes.set_xlim([-1.,1.])
        axes.set_ylim([-1.,1.])
        axes.set_xticks([])
        axes.set_yticks([])
        plt.title('V, iter = %d, m=%d' % (i,n))
        
        x = rvs3_rep[k, 0, :, i]
        y = rvs3_rep[k, 1, :, i]
        ax = fig.add_subplot(3*Nvals,I,(3*k+2)*I+j+1)
        ax.scatter(x, y, color='g')
        plot_A_b(A, b, xx, xlims, ylims)
        axes = plt.gca()
        axes.set_xlim([-1.,1.])
        axes.set_ylim([-1.,1.])
        axes.set_xticks([])
        axes.set_yticks([])
        plt.title('J, iter = %d, m=%d' % (i,n))
fig.subplots_adjust(hspace=.5)

# fig.tight_layout()
# plt.suptitle('Dikin vs Vaidya, Exp = %d, n = %d ' % (experiments,n), fontsize=20)
plt.show()
```
