```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import main_run
```

```python
np.random.seed(695829)
experiments = 100
iterations = 50

# Defining the Polytope
d = 2
r = 0.5
rep_max = 6


rvs_rep = np.zeros((rep_max+1, d, experiments, iterations))
rvs2_rep = np.zeros((rep_max+1, d, experiments, iterations))
rvs3_rep = np.zeros((rep_max+1, d, experiments, iterations))

for rep in range(0, rep_max, 2):
    print 'rep = %d ' % rep
    A, b = main_run.define_polytope(d, rep)
    rvs_rep[rep,:, :, :] = main_run.run_dikin_walk(experiments, iterations, A, b, r)
    rvs2_rep[rep,:, :, :] = main_run.run_vaidya_walk(experiments, iterations, A, b, r)
    rvs3_rep[rep,:, :, :] = main_run.run_john_walk(experiments, iterations, A, b, r)
    
```

```python
fig = plt.figure(figsize = [10,20])

for jk, k in enumerate([2, 4]):
    for j, i in enumerate(range(0, iterations, iterations/2-1)):
        x = rvs_rep[k, 1, :, i]
        y = rvs_rep[k, 0, :, i]
        ax = fig.add_subplot(6,3,3*jk+j+1)
        ax.scatter(x, y)
        axes = plt.gca()
        axes.set_xlim([-1.,1.])
        axes.set_ylim([-1.,1.])
        plt.title('Iter = %d; n = %d' % (i, 2**(k+2)))

        x = rvs2_rep[k, 1, :, i]
        y = rvs2_rep[k, 0, :, i]
        ax = fig.add_subplot(6,3,6+3*jk+j+1)
        ax.scatter(x, y, color='r')
        axes = plt.gca()
        axes.set_xlim([-1.,1.])
        axes.set_ylim([-1.,1.])
        plt.title('Iter = %d; n = %d' % (i, 2**(k+2)))

        x = rvs3_rep[k, 1, :, i]
        y = rvs3_rep[k, 0, :, i]
        ax = fig.add_subplot(6,3,12+3*jk+j+1)
        ax.scatter(x, y, color='g')
        axes = plt.gca()
        axes.set_xlim([-1.,1.])
        axes.set_ylim([-1.,1.])
        plt.title('Iter = %d; n = %d' % (i, 2**(k+2)))

plt.show()
```
