```python
np.random.seed(695829)
experiments = 1000
iterations = 1001

# Defining the Polytope
d = 2
r = 0.5

repetitions = np.array([0, 4, 9])
num_reps = len(repetitions)

dikin_rvs = np.zeros((num_reps, d, experiments, iterations))
vaidya_rvs = np.zeros((num_reps, d, experiments, iterations))
john_rvs = np.zeros((num_reps, d, experiments, iterations))
for index, rep in enumerate(repetitions):
    print 'repetition = %d ' % rep
    A, b = main_run.define_polytope(d, rep)
    dikin_rvs[index,:, :, :] = main_run.run_dikin_walk(experiments, iterations, A, b, r)
    vaidya_rvs[index,:, :, :] = main_run.run_vaidya_walk(experiments, iterations, A, b, r)
#     john_rvs[index,:, :, :] = main_run.run_john_walk(experiments, iterations, A, b, r)
uniform_rvs = 2*np.random.rand(experiments, d)-1
```
